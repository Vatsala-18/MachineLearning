```python
df.rename(columns = {'Modal Price (Rs/q)' : 'y', 'Date' : 'ds'},inplace = True) #df updated & modified here itself (inplace=True)
```

```python
#create a new dataframe only by taking y and ds
df_new = df[['ds','y']]  #slicing the columns
df_new
```

```python
df_new.info()
```

```python
#convert ds to datetime
#conver y to int 
```

```python
df_new['ds'] = pd.to_datetime(df_new['ds']) #convert to datetime format
```

```python
df_new['y'] = df_new['y'].astype(int)
```

```python


df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
#FBPROPHET
```

```python
#threshold - create an outlier 
import numpy as np
df_new['y'] = np.where(df_new['y']>2700, np.nan, df_new['y'])
```

```python
df_new
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
df_new.isnull().sum()
```

```python
#using fbprophet
import fbprophet
op = fbprophet.Prophet(changepoint_prior_scale=1) #makes the graph more robust and fits nicely
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods=365,freq='D') #gives forecast for next 365 days
```

```python
forecast = op.predict(forecast)  #predict new values after forecasting
```

```python

plt.figure(figsize=(25,10))
op.plot(forecast,xlabel = 'Date',ylabel ='Potato Price/Q')
plt.show()
```
